# **Cleaning Data**

**Duplikasi Data**

```python
df.duplicated().sum()
```

**Inkonsistensi Kategori**

```python
df['Kategori'].unique()
```

```python
df['Kategori'] = df['Kategori'].replace('elektronik', 'Elektronik')
```

```python
df['Kategori'].unique()
```

**Missing Values**

```python
df.isnull().sum()
```

```python
df['Produk'] = df['Produk'].fillna('Tidak Diketahui')
```

```python
df['Jumlah'] = df['Jumlah'].fillna(df['Jumlah'].median())
```

```python
df['Harga_Satuan'] = df['Harga_Satuan'].fillna(df['Harga_Satuan'].median())
```

```python
df.isnull().sum()
```

**Perbaikan Format Pelanggan**

```python
df['Pelanggan'] = df['Pelanggan'].fillna(df['Pelanggan'].mode()[0])
```

**Perbaikan Format Tanggal**

```python
df['Tanggal'] = df['Tanggal'].fillna(df['Tanggal'].mode()[0])
```

```python
df.loc[7, 'Tanggal'] = '11/6/2023'
```

```python
print(df['Tanggal'])
```

```python
df.loc[2, 'Tanggal'] = '2023-11-02'
df.loc[6, 'Tanggal'] = '2023-11-06'
df.loc[7, 'Tanggal'] = '2023-11-07'
df.loc[14, 'Tanggal'] = '2023-11-13'
```

```python
df['Tanggal'] = pd.to_datetime(df['Tanggal'])
```

```python
print(df['Tanggal'].dtype)
```

```python
print(df['Tanggal'])
```

```python
df.isnull().sum()
```

**Spasi Berlebih**

```python
df = df.apply(
    lambda x: x.str.strip() if x.dtype == "object" else x
)
```

**Outlier**

```python
df['Total_Bayar'].describe()
```

```python
Q1 = df['Total_Bayar'].quantile(0.25)
Q3 = df['Total_Bayar'].quantile(0.75)

IQR = Q3 - Q1

bawah = Q1 - 1.5 * IQR
atas = Q3 + 1.5 * IQR

print("Q1 =", Q1)
print("Q3 =", Q3)
print("IQR =", IQR)
print("Batas bawah =", bawah)
print("Batas atas =", atas)
```

```python
df[df['Total_Bayar'] < 0]
```

```python
df = df[
    (df['Total_Bayar'] >= 0) &
    (df['Total_Bayar'] <= atas)
]
```

```python
print(df.shape)
```

**Simpan Data Bersih**

```python
df.to_csv('penjualan_bersih.csv', index=False)
```

# **VISUALISASI DATA**

**Bar Chart**

```python
kota = df.groupby('Kota')['Total_Bayar'].sum()

plt.figure(figsize=(8,5))

kota.plot(kind='bar')

plt.title('Total Penjualan per Kota')
plt.xlabel('Kota')
plt.ylabel('Total Penjualan')

plt.show()
```

**Pie Chart**

```python
kategori = df.groupby('Kategori')['Total_Bayar'].sum()

plt.figure(figsize=(6,6))

plt.pie(
    kategori,
    labels=kategori.index,
    autopct='%1.1f%%'
)

plt.title('Proporsi Penjualan per Kategori')

plt.show()
```

**Line Chart**

```python
harian = df.groupby('Tanggal')['Total_Bayar'].sum()

plt.figure(figsize=(12,5))

plt.plot(
    harian.index,
    harian.values,
    marker='o'
)

plt.title('Tren Penjualan Harian')
plt.xlabel('Tanggal')
plt.ylabel('Total Penjualan')

plt.xticks(rotation=45)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# DATA UNTUK BAR CHART
kota = df.groupby('Kota')['Total_Bayar'].sum().sort_values(ascending=False)

# DATA UNTUK PIE CHART
kategori = df.groupby('Kategori')['Total_Bayar'].sum()

# DATA UNTUK LINE CHART
harian = df.groupby('Tanggal')['Total_Bayar'].sum()

# HASIL ANALISIS
fig = plt.figure(figsize=(16,10))

# JUDUL UTAMA
fig.suptitle(
    'Analisis Penjualan\nNovember - Desember 2023',
    fontsize=20,
    fontweight='bold'
)

# BAR CHART
ax1 = plt.subplot(2,2,1)

warna_bar = ['red','yellow','green','blue','purple']

bars = ax1.bar(
    kota.index,
    kota.values,
    color=warna_bar
)

ax1.set_title('Total Penjualan per Kota')
ax1.set_ylabel('Total Penjualan (Rp)')
ax1.grid(axis='y', linestyle='--', alpha=0.4)

for bar in bars:
    y = bar.get_height()
    ax1.text(
        bar.get_x()+bar.get_width()/2,
        y,
        f'Rp {y/1000000:.1f}M',
        ha='center',
        va='bottom',
        fontsize=9
    )

# PIE CHART
ax2 = plt.subplot(2,2,2)

ax2.pie(
    kategori,
    labels=kategori.index,
    autopct='%1.1f%%',
    explode=[0.05]*len(kategori),
    shadow=True
)

ax2.set_title('Proporsi Penjualan per Kategori')

# LINE CHART
ax3 = plt.subplot(2,1,2)

ax3.plot(
    harian.index,
    harian.values,
    marker='o',
    linewidth=2
)

ax3.fill_between(
    harian.index,
    harian.values,
    alpha=0.2
)

ax3.set_title('Tren Penjualan Harian')
ax3.set_xlabel('Tanggal')
ax3.set_ylabel('Total Penjualan (Rp)')
ax3.grid(True, linestyle='--', alpha=0.4)

plt.xticks(rotation=45)

plt.tight_layout()

import matplotlib.ticker as ticker

ax1.yaxis.set_major_formatter(
    ticker.FuncFormatter(
        lambda x, pos: f'Rp {x/1000000:.0f}M'
    )
)

ax3.yaxis.set_major_formatter(
    ticker.FuncFormatter(
        lambda x, pos: f'Rp {x/1000000:.0f}M'
    )
)

plt.show()
```
